```python
import sympy
from sympy.plotting import plot
```

```python
x = sympy.symbols('x')
```

```python
f_x = -12 * x**4 * sympy.sin(sympy.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
print(f_x)
```

```python
print(f_x.subs(x, 10.0))
```

```python
plot(f_x)
```

```python
diff_f_x = sympy.diff(f_x, x)
print(diff_f_x)
```

```python
for i in range(76506220, 76506230):
    print(i / 10000000, f_x.subs(x, -i / 10000000)) #промежуток для 1 корня
```
